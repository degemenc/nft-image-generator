```python
from PIL import Image 
from IPython.display import display 
import random
import json
import os
```

```python
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%

def getLayersAndTraits():
  path = "./trait-layers"
  layers = os.listdir(path)

  filtered_layers = []
  traits = {}
  for layer in layers:
    if layer[0] == ".":
      continue
    filtered_layers.append(layer)
    layer_traits = os.listdir("./trait-layers/" + layer)
    if layer not in traits.keys():
      traits[layer] = {
        "filenames": {},
        "assets": [],
        "weights": []
      }
    for lt in layer_traits:
      if lt[0] == ".":
        continue
      
      asset_and_weight = lt[:len(lt) - 4]
      split = asset_and_weight.split("__") #double __
      asset = split[0]
      weight = 100
      if len(split) > 1:
        weight = int(split[1])
      traits[layer]["filenames"][asset] = lt
      traits[layer]["assets"].append(asset)
      traits[layer]["weights"].append(weight)
  filtered_layers.sort()
  return [filtered_layers, traits]
  
LT = getLayersAndTraits()
print(LT)
layers = LT[0]
traits = LT[1]

headgear_one_of = ["layer_0050_hairs", "layer_0080_hat_under_headphones", "layer_0100_hat_over_headphones",]
```

```python
## Generate Traits

TOTAL_IMAGES = 200 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #
    selected_headgear = headgear_one_of[random.randint(0,2)]
    
    for layer in layers:
        if layer in headgear_one_of and layer != selected_headgear:
            continue
        print('layer: ' + str(layer))
        print('traits[layer]["assets"]: ' + str(traits[layer]["assets"]))
        print('traits[layer]["weights"]: ' + str(traits[layer]["weights"]))
        new_image[layer] = random.choices(traits[layer]["assets"], traits[layer]["weights"])[0]
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    print('new_trait_image: ' + str(new_trait_image))
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
```

```python
#### Generate Metadata for all Traits 
METADATA_FILE_NAME = './metadata/all-traits.json'; 
with open(METADATA_FILE_NAME, 'w') as outfile:
    json.dump(all_images, outfile, indent=4)
```

```python
#### Generate Images    
print('layers: ' + str(layers))
for item in all_images:
    
    com = None
    first = True
    for layer in layers:
        layerAsset = item.get(layer, None)
        if layerAsset == None:
            continue
        img = Image.open(f'./trait-layers/{layer}/{traits[layer]["filenames"][layerAsset]}').convert('RGBA')
        if first:
            com = img
            first = False
        else:
            com = Image.alpha_composite(com, img)

    #Convert to RGB
    rgb_im = com.convert('RGB')
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./images/" + file_name)
```
